```python
#get all courses from an Excel file
def courses(filename) :
    df_courses = pd.DataFrame()
    df_courses_final = pd.DataFrame()
    #to get the number of sheets
    document = xlrd.open_workbook(filename)
    nsheets = document.nsheets
    #to read and stock the first sheet
    xls = pd.ExcelFile(filename)
    #df_file_name = pd.read_excel(xls, 0)
    #for all the others sheets
    for i in range(0, nsheets, 1) :
        df = pd.read_excel(xls, i)
        df_courses = df_courses.append(df, ignore_index=True, sort=False)
    df_courses_final = df_courses_final.append(df_courses,ignore_index=True, sort=False)
    return df_courses_final
```

```python
def all_courses(files) :
    oue = pd.DataFrame()
    for i in files_xls :
        oue = oue.append(courses(i),ignore_index=True, sort=False)
    return oue
```

def create_entries(df):
    entries = []
    # Stopping case
    if df.shape[1] == 2:  # only 2 columns left
        for i in range(df.shape[0]):  # iterating on rows
            entries.append(
                {"name": df.iloc[i, 0],
                 df.columns[-1]: df.iloc[i, 1]}
            )
    # Iterating case
    else:
        values = set(df.iloc[:, 0])  # Getting the set of unique values
        for v in values:
            entries.append(
                {"name": v,
                 # reiterating the process but without the first column
                 # and only the rows with the current value
                 "children": create_entries(
                     df.loc[df.iloc[:, 0] == v].iloc[:, 1:]
                 )}
            )
    return entries

class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(MyEncoder, self).default(obj)
